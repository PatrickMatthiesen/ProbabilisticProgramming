```python
# examples of distribution plots
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd

np.random.seed(0)
```

```python
# Normal Distribution plots
# Mean of the distribution 
Mean = 100

# satndard deviation of the distribution
Standard_deviation  = 5
 
# size
size = 100000
 
# creating a normal distribution data
values = np.random.normal(Mean, Standard_deviation, size)
 
# plotting histograph
plt.hist(values, 100)
# plotting mean line
plt.axvline(values.mean(), color='k', linestyle='dashed', linewidth=2)
plt.show()
```

```python
# Uniform Distribution plots
# size
size = 100000

# creating a uniform distribution data
values = np.random.uniform(-10, 10, size)

# plotting histograph
plt.hist(values, 100)
# plotting mean line
plt.axvline(values.mean(), color='k', linestyle='dashed', linewidth=2)
plt.show()
```

```python
# Exponential Distribution plots
# size
size = 100000

# creating a exponential distribution data
values = np.random.exponential(3, size)

# plotting histograph
plt.hist(values, 100)
# plotting mean line
plt.axvline(values.mean(), color='k', linestyle='dashed', linewidth=2)
plt.show()
```

```python
# Log-Normal Distribution plots
# size
size = 10000

# creating a log-normal distribution data
values = np.random.lognormal(10, 1, size)

# plotting histograph
plt.hist(values, 100)
# plotting mean line
plt.axvline(values.mean(), color='k', linestyle='dashed', linewidth=2)
plt.show()
```

```python
# Poisson Distribution plots
# size
size = 100

# creating a poisson distribution data
values = np.random.poisson(3, size)

# plotting histograph
plt.hist(values, 100)
# plotting mean line
plt.axvline(values.mean(), color='k', linestyle='dashed', linewidth=2)
plt.show()
```
